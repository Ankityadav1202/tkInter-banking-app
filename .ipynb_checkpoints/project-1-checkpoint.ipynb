# Front end designing(presentation part)

```python
from tkinter import *
import time
from tkinter import messagebox
from tkinter.ttk import Combobox
import pymysql
```

```python
try:
    conobj=pymysql.connect(user="root",password="root",database="Banking")
    curobj=conobj.cursor()
    curobj.execute("create table accounts(acn_no int primary key auto_increment,acn_name varchar(30),acn_pass varchar(20),acn_email varchar(40),acn_mob int,acn_bal float,acn_type varchar(30),acn_opendate varchar(30)")
    curobj.execute("create table txns(acn_no int,txn_amt float,txn_type varchar(20),txn_updatebal float,txn_date varchar(30))")
    conobj.commit()
    print('tables created')
except:
    print('something went wrong,might be table already exists')
    conobj.close()

win=Tk()
win.state('zoomed')
win.configure(bg='powder blue')
win.resizable(width=False,height=False)


title_lbl=Label(win,text='Banking Automation',font=('arial',60,'bold','underline'),bg='powder blue')
title_lbl.pack()

date_lbl=Label(win,text=time.strftime('%d,%B,%y'),font=('arial',15,'bold'),bg='powder blue')
date_lbl.place(relx=.9,rely=.1)

time_lbl=Label(win,text=time.strftime('%I:%M:%S'),font=('arial',15,'bold'),bg='powder blue')
time_lbl.place(relx=.0,rely=.1)



def main_screen():
    frm=Frame(win)
    frm.configure(bg='pink')
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    
    def new_click():
        frm.destroy()
        openaccount_screen()
        
    def forgotpass_click():
        frm.destroy()
        forgotpass_screen()
    
    def clear():
        acn_entry.delete(0,'end')
        pass_entry.delete(0,'end')
        acn_entry.focus()   
        
    def loginclick():
            a=acn_entry.get()
            p=pass_entry.get()
            if len(a)==0 or len(p)==0:
                messagebox.showerror('ERROR','Invalid Account & Password')
            
            else:
                frm.destroy()
                welcome_screen()        
        
    
               
    
    acn_lbl=Label(frm,text='ACN',font=('arial',15,'bold'),bg='pink',fg='blue')
    acn_lbl.place(relx=.4,rely=.1)
    
    acn_entry=Entry(frm,font=('',15,'bold'),bd=5)
    acn_entry.place(relx=.45,rely=.1)
    
    pass_lbl=Label(frm,text='Password',font=('arial',15,'bold'),bg='pink',fg='blue')
    pass_lbl.place(relx=.36,rely=.2)
    
    pass_entry=Entry(frm,font=('',15,'bold'),bd=5,show='*')#show keyword to hide entry and show what you mention
    pass_entry.place(relx=.45,rely=.2)
    
    login_btn=Button(frm,command=loginclick,text='Login',font=('arial',15,'bold'),bd=3,bg='powder blue')
    login_btn.place(relx=.45,rely=.3)
    
    clear_btn=Button(frm,text='Clear',font=('arial',15,'bold'),bd=3,bg='powder blue',command=clear)
    clear_btn.place(relx=.58,rely=.3)
    
    forgot_btn=Button(frm,command=forgotpass_click,text='Forgot Pass',width=15,font=('arial',15,'bold'),bd=3,bg='powder blue')
    forgot_btn.place(relx=.47,rely=.4)
    
    new_btn=Button(frm,command=new_click,text='Open new account',width=15,font=('arial',15,'bold'),bd=3,bg='powder blue')
    new_btn.place(relx=.47,rely=.5)

def openaccount_screen():
    frm=Frame(win)
    frm.configure(bg='pink')
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    
    def back_btn():
        frm.destroy()
        main_screen()
        
    def newac_db():
        conobj=pymysql.connect(user="root",password="root",database="Banking")
        curobj=conobj.cursor()
        
        name=name_entry.get()
        pwd=pass_entry.get()
        email=email_entry.get()
        mob=mob_entry.get()
        acntype=cmb_entry.get()
        bal=0
        opendate=str(time.ctime())
        curobj.execute("insert into accounts(acn_name,acn_pass,acn_email,acn_mob,acn_bal,acn_type,acn_opendate) values(%s,%s,%s,%s,%s,%s,%s)",(name,pwd,email,mob,bal,acntype,opendate))
        conobj.commit()
        curobj.close()
        
        curobj=conobj.cursor()
        curobj.execute("select max(acn_no) from accounts")
        tup=curobj.fetchone()
        conobj.close()
        
        messagebox.showinfo("New Account",f"Your Account is open with ACN={tup[0]}")
        
        
        
        
    
    def clear():
        name_entry.delete(0,'end')
        pass_entry.delete(0,'end')
        email_entry.delete(0,'end')
        mob_entry.delete(0,'end')
        cmb_entry.delete(0,'end')
        name_entry.focus()   
        
    back_button=Button(frm,command=back_btn,text='back',font=('',14,'bold'),bd=3,bg='powder blue')
    back_button.place(relx=0,rely=0)
    
    name_lbl=Label(frm,text='Name',font=('arial',15,'bold'),bg='pink',fg='blue')
    name_lbl.place(relx=.4,rely=.1)
    name_lbl.focus
    
    name_entry=Entry(frm,font=('',15,'bold'),bd=5)
    name_entry.place(relx=.45,rely=.1)
    
    pass_lbl=Label(frm,text='pass',font=('arial',15,'bold'),bg='pink',fg='blue')
    pass_lbl.place(relx=.4,rely=.2)
    
    pass_entry=Entry(frm,font=('',15,'bold'),bd=5,show='*')
    pass_entry.place(relx=.45,rely=.2)
    
    email_lbl=Label(frm,text='Email',font=('arial',15,'bold'),bg='pink',fg='blue')
    email_lbl.place(relx=.4,rely=.3)
    
    email_entry=Entry(frm,width=25,font=('',15,'bold'),bd=5)
    email_entry.place(relx=.45,rely=.3)
    
    mob_lbl=Label(frm,text='Mob',font=('arial',15,'bold'),bg='pink',fg='blue')
    mob_lbl.place(relx=.4,rely=.4)
    
    mob_entry=Entry(frm,font=('',15,'bold'),bd=5)
    mob_entry.place(relx=.45,rely=.4)
    
    cmb_lbl=Label(frm,text='Type',font=('arial',15,'bold'),bg='pink',fg='blue')
    cmb_lbl.place(relx=.4,rely=.5)
    
    cmb_entry=Combobox(frm,font=('',15,'bold'),values=['Saving','Current','FD'])
    cmb_entry.current(0)
    cmb_entry.place(relx=.45,rely=.5)
    
    submit_button=Button(frm,command=newac_db,text='submit',font=('',15,'bold'),bd=3,bg='powder blue')
    submit_button.place(relx=.45,rely=.6)
    
    clear_button=Button(frm,text='clear',font=('',15,'bold'),bd=3,bg='powder blue',command=clear)
    clear_button.place(relx=.58,rely=.6)
    
def forgotpass_screen():
    frm=Frame(win)
    frm.configure(bg='pink')
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    
    def back_btn():
        frm.destroy()
        main_screen()
        
    def clear():
        name_entry.delete(0,'end')
        email_entry.delete(0,'end')
        mob_entry.delete(0,'end')
        name_entry.focus()       
        
    back_button=Button(frm,command=back_btn,text='back',font=('',14,'bold'),bd=3,bg='powder blue')
    back_button.place(relx=0,rely=0)
    
    name_lbl=Label(frm,text='Name',font=('arial',15,'bold'),bg='pink',fg='blue')
    name_lbl.place(relx=.4,rely=.1)
    name_lbl.focus
    
    name_entry=Entry(frm,font=('',15,'bold'),bd=5)
    name_entry.place(relx=.45,rely=.1)
    
    email_lbl=Label(frm,text='Email',font=('arial',15,'bold'),bg='pink',fg='blue')
    email_lbl.place(relx=.4,rely=.2)
    
    email_entry=Entry(frm,width=25,font=('',15,'bold'),bd=5)
    email_entry.place(relx=.45,rely=.2)
    
    mob_lbl=Label(frm,text='Mob',font=('arial',15,'bold'),bg='pink',fg='blue')
    mob_lbl.place(relx=.4,rely=.3)
    
    mob_entry=Entry(frm,font=('',15,'bold'),bd=5)
    mob_entry.place(relx=.45,rely=.3)
    
    submit_button=Button(frm,text='submit',font=('',15,'bold'),bd=3,bg='powder blue')
    submit_button.place(relx=.45,rely=.4)
    
    clear_button=Button(frm,text='clear',font=('',15,'bold'),bd=3,bg='powder blue',command=clear)
    clear_button.place(relx=.58,rely=.4)
    
def welcome_screen():
    frm=Frame(win)
    frm.configure(bg='pink')
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    
    def logout_btn():
        res=messagebox.askyesno('logout','Do you want to logout?')
        if res==True:
            frm.destroy()
            main_screen() 
            
    def update_screen():
        ifrm=Frame(frm,highlightthickness=5,highlightbackground='black')
        ifrm.configure(bg='white')
        ifrm.place(relx=.2,rely=.1,relwidth=.6,relheight=.6)
        
        heading_lbl=Label(ifrm,text='This is update profile screen ',font=('arial',15,'bold'),bg='white',fg='purple')
        heading_lbl.pack()
        
    def balance_screen():
        ifrm=Frame(frm,highlightthickness=5,highlightbackground='black')
        ifrm.configure(bg='white')
        ifrm.place(relx=.2,rely=.1,relwidth=.6,relheight=.6)
        
        heading_lbl=Label(ifrm,text='This is check Balance screen ',font=('arial',15,'bold'),bg='white',fg='purple')
        heading_lbl.pack() 
        
    def deposit_screen():
        ifrm=Frame(frm,highlightthickness=5,highlightbackground='black')
        ifrm.configure(bg='white')
        ifrm.place(relx=.2,rely=.1,relwidth=.6,relheight=.6)
        
        heading_lbl=Label(ifrm,text='This is deposit screen ',font=('arial',15,'bold'),bg='white',fg='purple')
        heading_lbl.pack()
        
        amt_lbl=Label(ifrm,text='Amt',font=('arial',17,'bold'),bg='white',fg='blue')
        amt_lbl.place(relx=0.3,rely=.2)
        
        amt_entry=Entry(frm,width=25,font=('',15,'bold'),bd=5,bg='pink')
        amt_entry.place(relx=.45,rely=.23)
        
        submit_button=Button(frm,text='submit',font=('',15,'bold'),bd=3,bg='powder blue')
        submit_button.place(relx=.5,rely=.35)
    
        
    def withdraw_screen():
        ifrm=Frame(frm,highlightthickness=5,highlightbackground='black')
        ifrm.configure(bg='white')
        ifrm.place(relx=.2,rely=.1,relwidth=.6,relheight=.6)
        
        heading_lbl=Label(ifrm,text='This is withdraw screen ',font=('arial',15,'bold'),bg='white',fg='purple')
        heading_lbl.pack()
        
        amt_lbl=Label(ifrm,text='Amt',font=('arial',17,'bold'),bg='white',fg='blue')
        amt_lbl.place(relx=0.3,rely=.2)
        
        amt_entry=Entry(frm,width=25,font=('',15,'bold'),bd=5,bg='pink')
        amt_entry.place(relx=.45,rely=.23)
        
        submit_button=Button(frm,text='submit',font=('',15,'bold'),bd=3,bg='powder blue')
        submit_button.place(relx=.5,rely=.35)
        
    def transfer_screen():
        ifrm=Frame(frm,highlightthickness=5,highlightbackground='black')
        ifrm.configure(bg='white')
        ifrm.place(relx=.2,rely=.1,relwidth=.6,relheight=.6)
        
        heading_lbl=Label(ifrm,text='This is transfer screen ',font=('arial',15,'bold'),bg='white',fg='purple')
        heading_lbl.pack()
        
        amt_lbl=Label(ifrm,text='Amt',font=('arial',17,'bold'),bg='white',fg='blue')
        amt_lbl.place(relx=0.3,rely=.2)
        
        amt_entry=Entry(frm,width=25,font=('',15,'bold'),bd=5)
        amt_entry.place(relx=.45,rely=.23)
        
        submit_button=Button(frm,text='submit',font=('',15,'bold'),bd=3,bg='powder blue')
        submit_button.place(relx=.5,rely=.4)
        
        to_lbl=Label(ifrm,text='to',font=('arial',17,'bold'),bg='white',fg='blue')
        to_lbl.place(relx=0.3,rely=.37)
        
        to_entry=Entry(frm,width=25,font=('',15,'bold'),bd=5)
        to_entry.place(relx=.45,rely=.33)
        
        
        
    
    
    
        
    logout_button=Button(frm,command=logout_btn,text='Logout',font=('',14,'bold'),bd=3,bg='powder blue')
    logout_button.place(relx=.94,rely=0)
    
    mob_lbl=Label(frm,text='Welcome,..',font=('arial',15,'bold'),bg='pink',fg='blue')
    mob_lbl.place(relx=0,rely=0)
    
    update_button=Button(frm,command=update_screen,text='update profile',width=15,font=('',14,'bold'),bd=3,bg='powder blue')
    update_button.place(relx=0,rely=.1)
    
    bal_button=Button(frm,command=balance_screen,text='Check balance',width=15,font=('',14,'bold'),bd=3,bg='powder blue')
    bal_button.place(relx=0,rely=.2)
    
    deposit_button=Button(frm,command=deposit_screen,text='Deposit',width=15,font=('',14,'bold'),bd=3,bg='powder blue')
    deposit_button.place(relx=0,rely=.3)
    
    withdraw_button=Button(frm,command=withdraw_screen,text='Withdraw',width=15,font=('',14,'bold'),bd=3,bg='powder blue')
    withdraw_button.place(relx=0,rely=.4)
    
    transfer_button=Button(frm,command=transfer_screen,text='transfer',width=15,font=('',14,'bold'),bd=3,bg='powder blue')
    transfer_button.place(relx=0,rely=.5)
    
   
    
    
main_screen()                
win.mainloop()                
```
